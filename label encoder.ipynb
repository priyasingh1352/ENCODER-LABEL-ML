# Label Encoder

```python
import pandas as pd
```

```python
df = pd.DataFrame({"name":["Vivek","Mayank","Lucky","Sangeeta","Lucky","Monica","Priya","Arpita","Sonali","Monica","Leena","Lucky","Madhu"]})
```

```python
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit_transform(df["name"])
```

```python
df["en_ name"] = le.fit_transform(df["name"])
```

```python
df
```

```python
new_data = df
new_data.to_csv("Label_Encoder.csv",index=False)
```
